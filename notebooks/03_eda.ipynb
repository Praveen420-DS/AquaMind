```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.figure(figsize=(6,4))
df=pd.read_csv("../data/raw/water_potability.csv")
# ==========================================
# Visualization 1: Distribution of Potability
# Purpose:
# To understand the class distribution of the target variable.
# ==========================================
sns.countplot(data=df,x="Potability")
plt.title("Distribution of water Potability")
plt.xlabel("Potability")
plt.ylabel("count")
plt.show()
```

```python
# ==========================================
# Visualization 2: Histograms
# Purpose:
# To understand the distribution of each numerical feature.
# ==========================================

df.hist(figsize=(18,12), bins=20)

plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()

plt.show()
```

```python
# ==========================================
# Visualization 3: Box Plots
# Purpose:
# To identify outliers and understand the spread of each feature.
# ==========================================

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18, 10))

for i, column in enumerate(df.columns[:-1], 1):   # Exclude Potability
    plt.subplot(3, 3, i)
    sns.boxplot(y=df[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
Q1 = df["ph"].quantile(0.25)
Q3 = df["ph"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower)
print("Upper Bound:", upper)

outliers = df[(df["ph"] < lower) | (df["ph"] > upper)]

print("Number of Outliers:", len(outliers))
```

```python
# ==========================================
# Visualization 4: Correlation Heatmap
# Purpose:
# To understand relationships between features.
# ==========================================

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()
```
